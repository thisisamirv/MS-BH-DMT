```r
install.packages("dplyr")
install.packages("meta")
if (!require("devtools")) {
  install.packages("devtools")
}
devtools::install_github("MathiasHarrer/dmetar")
```

```r
library(meta)
library(dplyr)
library(dmetar)
```

```r
#Load data
Data <- readxl::read_excel("Data.xlsx")
Data <- Data[order(Data$`Study ID`) , ]
GeneralData <- Data[order(Data$`Study ID`) , c("Study ID", "N-I" , "N-P", "T1-V-I", "T1-V-I-SD", "T1-V-P", "T1-V-P-SD", "T1-N-I", "T1-N-I-SD", "T1-N-P", "T1-N-P-SD")]
```

```r
Data1 <- Data[-c(9), ]
Data1
GeneralData1 <- GeneralData[-c(3, 7, 10), ]
```

```r
#Main synthesis 1
MainSynthesis1 <- metacont(n.e = GeneralData1$'N-I', mean.e = GeneralData1$'T1-V-I', sd.e = GeneralData1$'T1-V-I-SD', n.c = GeneralData1$'N-P', mean.c = GeneralData1$'T1-V-P', sd.c = GeneralData1$'T1-V-P-SD', studlab = GeneralData1$'Study ID', text.w.random="Weight")
print(MainSynthesis1)
```

```r
#Main synthesis 2
MainSynthesis2 <- metacont(n.e = GeneralData$'N-I', mean.e = GeneralData$'T1-N-I', sd.e = GeneralData$'T1-N-I-SD', n.c = GeneralData$'N-P', mean.c = GeneralData$'T1-N-P', sd.c = GeneralData$'T1-N-P-SD', studlab = GeneralData$'Study ID', text.w.random="Weight")
print(MainSynthesis2)
```

```r
#Create forest plot 1
png(file = "./Main Synthesis 1.png", width = 4000, height = 1200, res = 300)
forest.meta(MainSynthesis1, test.overall = TRUE, fixed = FALSE, calcwidth.tests = TRUE, test.overall.random = TRUE, leftlabs = c("Study ID", "N", "Mean", "SD", "N", "Mean", "SD"), colgap.left = "0.5cm", just = "center", lab.e = "DMT", lab.c = "Placebo", xlim = c(-1500 , 200), col.inside = "black", col.square = "red", col.diamond = "black", digits = 1, digits.mean = 1, digits.sd = 1)
while (!is.null(dev.list()))
dev.off()
```

```r
#Bias Graphs
BiasCharts <- Data1 %>% 
  select( `Study ID` , 
          "1.Bias arising from the randomization process" = `RoB1` ,
          "2.Bias due to deviations from intended interventions" = `RoB2`, 
          "3.Bias due to missing outcome data" = `RoB3`, 
          "4.Bias in measurement of the outcome" = `RoB4` ,
          "5.Bias in selection of the reported result" = `RoB5`)

Bias_df<- as.data.frame(BiasCharts)

png(filename = "./bias.png", width = 1000, height = 250)
rob.summary(Bias_df, name.high = "High", name.unclear = "Unclear", name.low = "Low")
dev.off()

png(filename = "./biasTable.png", width = 400 , height = 1200)
rob.summary(Bias_df, studies = Bias_df$`Study ID` , name.high="High", name.unclear="Unclear", name.low="Low", table = T)
dev.off()

png(filename = "./biasTable2.png", width = 400 , height = 1200)
rob.summary(Bias_df, studies = Bias_df$`Study ID` , name.high="High", name.unclear="Unclear", name.low="Low", table = T)
dev.off()
```
