```r
install.packages("dplyr")
install.packages("meta")
if (!require("devtools")) {
  install.packages("devtools")
}
devtools::install_github("MathiasHarrer/dmetar")
```

```r
library(meta)
library(dplyr)
library(dmetar)
```

```r
#Load data
Data <- readxl::read_excel("Data.xlsx")
Data <- Data[order(Data$`Study ID`) , ]
GeneralData <- Data[order(Data$`Study ID`) , c("N-I" , "N-P", "T1-V-I", "T1-V-I-SD", "T1-V-P", "T1-V-P-SD", "T1-N-I", "T1-N-I-SD", "T1-N-P", "T1-N-P-SD")]
```

```r
#Main synthesis 1
MainSynthesis1 <- metacont(n.e = GeneralData$'N-I', mean.e = GeneralData$'T1-V-I', sd.e = GeneralData$'T1-V-I-SD', n.c = GeneralData$'N-P', mean.c = GeneralData$'T1-V-P', sd.c = GeneralData$'T1-V-P-SD', studlab = GeneralData$'Study ID', text.w.random="Weight")
print(MainSynthesis1)
```

```r
#Main synthesis 2
MainSynthesis2 <- metacont(n.e = GeneralData$'N-I', mean.e = GeneralData$'T1-N-I', sd.e = GeneralData$'T1-N-I-SD', n.c = GeneralData$'N-P', mean.c = GeneralData$'T1-N-P', sd.c = GeneralData$'T1-N-P-SD', studlab = GeneralData$'Study ID', text.w.random="Weight")
print(MainSynthesis2)
```

```r
#Create forest plot 1
png(file = "./Main Synthesis 1.png", width = 3500, height = 2100, res = 300)
forest.meta(MainSynthesis1, test.overall = TRUE, calcwidth.tests = TRUE, leftlabs = c("Study ID" , "Sample Size"), test.overall.random = TRUE, prediction = TRUE, col.square = "red", col.diamond = "black")
while (!is.null(dev.list()))
dev.off()
```
